```python
import pandas as pd

def new_csv(key):
    df = pd.DataFrame()
    for i in dict_list[key]:
        file_path = f'/content/drive/MyDrive/KOBERT/self_care/{i}.csv'
        temp_df = pd.read_csv(file_path, encoding='utf-8')
        df = pd.concat([df, temp_df])
    output_path = f'/content/drive/MyDrive/KOBERT/self_care/real/{key}.csv'
    df.to_csv(output_path, encoding='utf-8', index=False)
```

```python
for keys in dict_list.keys():
    new_csv(keys)
```
